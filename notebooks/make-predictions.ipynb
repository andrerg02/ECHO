```python
import sys
import os
# Add project root to path to import utils
sys.path.append(os.path.abspath('..'))

import torch
import lightning as L
from torch_geometric.loader import DataLoader
from utils import get_dataset
from utils.litmodels import LitGraphNN

# 1. Configuration
task = 'diam'  # Ensure this matches the trained model
checkpoint_path = "/your/path/to/checkpoint.ckpt"  # Directory where checkpoints are stored

print(f"Using checkpoint: {checkpoint_path}")
```

```python
# 2. Load Dataset
print(f"Loading dataset for task: {task}")
data_train, data_val, data_test, num_feat, num_class = get_dataset(
    root="../data",
    task=task
)

# 3. Create DataLoader
predict_loader = DataLoader(data_test, batch_size=512, shuffle=False, num_workers=4)

# 4. Load Model
print("Loading model...")
model = LitGraphNN.load_from_checkpoint(checkpoint_path)
model.eval()
```

```python
# 5. Make Predictions
print("Making predictions...")
trainer = L.Trainer(accelerator="gpu" if torch.cuda.is_available() else "cpu", enable_progress_bar=True)
test_metrics = trainer.test(model, dataloaders=predict_loader)
```
